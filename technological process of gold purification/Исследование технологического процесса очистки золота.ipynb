 ### Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака `rougher.output.recovery`. Найдите MAE между вашими расчётами и значением признака. Опишите выводы.

```python
#Напишем функцию для проверки эффективности обогащение 
def recovery (data):
    c = data['rougher.output.concentrate_au']
    f = data['rougher.input.feed_au']
    t = data['rougher.output.tail_au']
    return (c*(f-t))/(f*(c-t)) * 100
```

```python
#Найдем MAE между нашими расчётами и значением признака 
mean_absolute_error(df_train['rougher.output.recovery'], recovery(df_train))
```

```python
recovery(df_train)
```

```python
#проверим на адекватность значение MAE подставив вместо расчетов медиангое значение.
predicted_valid = pd.Series(df_train['rougher.output.recovery'].median(), index=df_train.index) 
mean_absolute_error(df_train['rougher.output.recovery'], recovery(df_train)) \
< mean_absolute_error(df_train['rougher.output.recovery'], predicted_valid)
```

В описании таблиц мы узнали, что соседние по времени параметры часто похожи. Поэтому заменим пропуски вышестоящим значением

```python
#напишем функцию заполняющую пропуски
def miss_columns_in(data):
    for i in data.columns:
        if data[i].isnull().sum() > 0:
            data[i].fillna(method='ffill', inplace=True)
```

### Посмотрите, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки. Опишите выводы.

Выделимм следующие этапы очистки:
1. подача на вход на этапе rougher; 
2. концентрат на этапе rougher;
3. концентрат на этапе primary_cleaner;
4. концентрат final.

На этапе srcondary_cleaner нет концентрата т.к. это последний этап очистки.

```python
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(df_full['rougher.input.feed_au'], label="подача золота на вход на этапе флотации", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df_full['rougher.output.concentrate_au'], label="Концентрация золота после флотации", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df_full['primary_cleaner.output.concentrate_au'], label="Концентрация золота после первичной очистки", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df_full['final.output.concentrate_au'], label="Финальная концентрация золота", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.set_palette(sns.color_palette('pastel'))

plt.ylim(0, 0.35)


# Decoration
plt.title('Концентрация золота на различных этапах очистки', fontsize=16)
plt.legend()
plt.show()
```

после каждого последовательного этапа графически можно увидеть, что концентрация золота увеличивается

```python
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(df_full['rougher.input.feed_ag'], label="подача серебра на вход на этапе флотации", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df_full['rougher.output.concentrate_ag'], label="Концентрация серебра после флотации", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df_full['primary_cleaner.output.concentrate_ag'], label="Концентрация серебра после первичной очистки", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df_full['final.output.concentrate_ag'], label="Финальная концентрация серебра", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.set_palette(sns.color_palette('pastel'))
plt.ylim(0, 0.35)


# Decoration
plt.title('Концентрация серебра на различных этапах очистки', fontsize=16)
plt.legend()
plt.show()
```

в Данном графике можем наблюдать обратную тенденцию. Концентрация серебра на финальном этапе меньше, чем на других этапах.  

```python
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(df_full['rougher.input.feed_pb'], label="подача свинца на вход на этапе флотации", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df_full['rougher.output.concentrate_pb'], label="Концентрация свинца после флотации", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df_full['primary_cleaner.output.concentrate_pb'], label="Концентрация свинца после первичной очистки", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df_full['final.output.concentrate_pb'], label="Финальная концентрация свинца", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
plt.ylim(0, 0.45)
sns.set_palette(sns.color_palette('pastel'))


# Decoration
plt.title('Концентрация свинца на различных этапах очистки', fontsize=16)
plt.legend()
plt.show()
```

Данный график поинтереснее. Концентрация свинца при подаче на вход на этапе флотации меньше. С каждым этапом концентрация увеличивается. После первичной обработки и в финальной концентрация примерно одинаковая

## Модель

### Напишите функцию для вычисления итоговой sMAPE.

```python
def smape(target, pred):
    return 2 * np.mean(np.abs(pred - target) / (np.abs(pred) + np.abs(target))) * 100
```

```python
def final_sMAPE(sMAPE_rougher,sMAPE_final):
    return (0.25*sMAPE_rougher + 0.75*sMAPE_final)
```
